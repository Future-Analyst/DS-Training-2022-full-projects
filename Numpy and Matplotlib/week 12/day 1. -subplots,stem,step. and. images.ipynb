## stem plot and Step Plots

## Subplots

```python
x=np.linspace(-20,20,1000)
```

```python
sine=np.sin(x)
cos = np.cos(x)
plt.plot(x,sine)
plt.plot(x,cos)
```

```python
plt.subplots(2)
```

```python
fig,ax=plt.subplots(2)
```

```python
fig
```

```python
ax
```

```python
ax[0].plot(x,sine)
ax[1].plot(x,cos)
fig

```

```python
fig,ax=plt.subplots(1,2)
```

```python
ax[0].plot(x,sine)
ax[1].plot(x,cos)
ax[0].set_title("Sine Graph")
fig
```
