```python
def getScores(X_train,X_test,y_train,y_test,Classifier=LinearRegression,**kwargs):
    model=Classifier(**kwargs)
    model.fit(X_train,y_train)
    training_score=model.score(X_train,y_train)
    testing_score=model.score(X_test,y_test)
    print(f"Training Score:{training_score}\nTesting Score:{testing_score}")
```

```python
getScores(X_train,X_test,y_train,y_test)
```

```python
def transformToPeriod(time):
    hrs,mins=time.split(":")
    hrs=int(hrs)
    mins=int(mins)
    totalMins=hrs*60+mins
    if(totalMins<=719):
        return "Morning"
    elif totalMins<=959:
        return "Afternoon"
    return "Evening"
    
```

```python
transformToPeriod("16:00")
```

```python
df.Time.apply(transformToPeriod)
```

```python
time_period=X["Time"]=df.Time.apply(transformToPeriod)
```
